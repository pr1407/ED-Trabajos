```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
G.add_node(1)
G.add_node(2)
G.add_node(3)
```

```python
G.add_edge(1,2)
G.add_edge(2,3)
G.add_edge(1,3)
```

```python
print(G.nodes)
```

```python
print(G.edges)
```

```python
nx.draw(G)
```

```python
plt.show()
```
